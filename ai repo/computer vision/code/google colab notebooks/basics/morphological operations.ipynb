```python
import cv2
import numpy as np

img = cv.imread('abc.jpg', cv.IMREAD_GRAYSCALE)

# Create a rectangular structuring element of size 5x5
se_rect = cv2.getStructuringElement(cv2.MORPH_RECT, (5, 5))
print(f"Rectangular element:\n{se_rect}")

eroded = cv2.erode(img, se_rect)
dilated = cv2.dilate(img, se_rect)
opened = cv2.dilate(eroded, se_rect)
closed = cv2.erode(dilated, se_rect)

# Key Morphological Operations
# Erosion: Shrinks foreground objects and is useful for removing noise or disconnecting objects. It is performed using cv2.erode(src, kernel, ...).
# Dilation: Expands foreground objects and helps fill small holes or connect broken parts. It is performed using cv2.dilate(src, kernel, ...).
# Opening: An erosion followed by a dilation that removes small noise while preserving larger objects.
# Closing: A dilation followed by an erosion that fills small holes and closes gaps between components.
# Morphological Gradient: The difference between dilation and erosion, resulting in an object outline.
# Top Hat (White Hat): The difference between the original image and its opening, highlighting bright regions on dark backgrounds.
# Black Hat: The difference between the closing of the image and the original image, highlighting dark regions on light backgrounds.
# Structuring Element
# The kernel defines the neighborhood for operations. It can be created with NumPy (np.ones()) or cv2.getStructuringElement(). OpenCV supports rectangular, cross-shaped, and elliptical kernels.
```
