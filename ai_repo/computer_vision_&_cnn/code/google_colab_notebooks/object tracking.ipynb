```python
from ultralytics import YOLO

model = YOLO("yolov8n.pt")
```

```python
results = model.track(
    source=f"{project_path}/video.mp4",
    tracker="botsort.yaml",  # or bytetrack.yaml
    conf=0.3,
    iou=0.5,
    device=0,
    save=True,
    project=project_path,
    name="predictions"
)
```
